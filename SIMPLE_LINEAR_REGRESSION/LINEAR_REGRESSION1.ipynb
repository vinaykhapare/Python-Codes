LinearRegression on defined data

```python
#importing packages(library)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)     # Hours studied, The reshape method is used to change the shape of the array. The argument
#-1 tells NumPy to automatically determine the size of that dimension based on the size of the original array and the other dimensions 
#specified. Here, -1 lets NumPy calculate the number of rows needed, while 1 specifies that we want a single column.

Y = np.array([50, 56, 59, 65, 72]) # Exam scores

#Create and fit the model
model = LinearRegression()
model.fit(X, Y)


#coeffitients
beta_0 = model.intercept_
beta_1 = model.coef_[0]


print(f"Intercept (beta_0): {beta_0}")
print(f"Slope (beta_1): {beta_1}")

# Predict
Y_pred = model.predict(X)
print(Y_pred)

new_x = np.array([6, 7 ,8]).reshape(-1, 1)
new_y_predit = model.predict(new_x)
print(new_y_predit)


plt.scatter(X, Y, color='blue', label='Actual scores')
plt.plot(X, Y_pred, color='red', label='Fitted line')  # Add regression line
plt.scatter(new_x, new_y_predit, color= 'blue', label ='New prediction for X', marker='x')

plt.plot(new_x, new_y_predit, color= 'black', label='Fitted line')  # Add regression line

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Hours Studied vs Exam Score')
plt.legend()
plt.show()
```
