Line Plot  without csv file


```python
import matplotlib.pyplot as plt

x= [1,2,3,4]
y= [5,10,15,20]

# plt.plot(x,y,) #Simple

plt.plot(x, y, color='blue', linestyle='--', linewidth=2, marker='*',markersize='15', markerfacecolor='yellow')
# Create a line plot with customization

plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
plt.scatter(x, y, color='blue',  linewidth=2 )
plt.title('Scatter Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 5, 2, 4]

plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.bar(categories, values)

plt.show()
```

```python
import pandas as pd

# Sample data in a DataFrame
data = {
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Sales': [250, 100, 180, 400, 450]
}

df = pd.DataFrame(data)
print(df)

# Create a line plot using DataFrame

df.plot(x='Year', y='Sales' , kind='line')

# Add title and labels

plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')

# Show plot
plt.show()
```

```python
# Importing Seaborn

import seaborn as sns
# seaborn is also like matplotlib 
# used for plot,
# data visualization 
# some extra features

# Load built-in dataset --> the 'tips' dataset

tips = sns.load_dataset("tips")
# print(tips)

# scatter-plot or line 
print(sns.scatterplot(x="total_bill", y="tip", data=tips))
```

```python
# sns.histplot(tips["total_bill"], bins=10)
sns.histplot(tips["total_bill"],bins=10)

# print(tips.head(10))
plt.show()
```

```python
# Box Plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# Bar Plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# Heatmap
# Generate a matrix of random numbers
import numpy as np
data = np.random.rand(10,10)
print(data)

# Create the heatmap
sns.heatmap(data )
plt.show()
```

Pair Plot
To visualize pairwise relationships in a dataset, you can use pairplot

```python
sns.pairplot(tips)
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.histplot, "total_bill")
plt.show()
```
